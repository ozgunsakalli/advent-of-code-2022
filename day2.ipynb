```python
with open('input-day2.txt', 'rt') as f:
    g = f.read()
```

```python
# Strategy guide
g[:100]
```

```python
g_list = g.split('\n')
g_list[:10]
```

```python
g_tuples = [tuple(n.split(' ')) for n in g_list]
g_tuples[:10]
```

```python
score = 0

for item in g_tuples:

    # score for shape
    if 'X' in item:
        score += 1
    elif 'Y' in item:
        score += 2
    elif 'Z' in item:
        score += 3

    # score for outcome
    if item in (('A', 'X'), ('B', 'Y'), ('C', 'Z')):
        # draw
        score += 3
    elif item in (('A', 'Y'), ('B', 'Z'), ('C', 'X')):
        # win
        score += 6
    else:
        # loss
        pass
```

```python
# part 1 answer
score
```

Part 2

```python
score = 0

for item in g_tuples:

    if item == '':
        continue

    to_add = 0
    # score for outcome
    if 'X' in item:
        pass     # loss
    elif 'Y' in item:
        to_add += 3      # draw
    elif 'Z' in item:
        to_add += 6      # win

    # score for shape
    if item in (('B', 'X'), ('A', 'Y'), ('C', 'Z')):
        # rock
        to_add += 1
    elif item in (('A', 'Z'), ('B', 'Y'), ('C', 'X')):
        # paper
        to_add += 2
    elif item in (('A', 'X'), ('B', 'Z'), ('C', 'Y')):
        # scissors
        to_add += 3
    
    score += to_add

    print(item, to_add)
```
