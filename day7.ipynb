```python
import copy

class Interpreter:

    def __init__(self, log):
        self.log = copy.deepcopy(log)
        self.cursor = 0
        self.tree = {'/': {}}
        self.current_dir = self.tree['/']
        self.current_dir_list = ['/']
    
    def proceed(self):
        
        while self.cursor < len(self.log):
            self.current_command = self.log[self.cursor]
            # print(self.current_command)
            if self.current_command.startswith('$ cd'):
                self.chdir()
            elif self.current_command == ('$ ls'):
                pass
            elif self.current_command.startswith('dir'):
                _, dirname = self.current_command.split(' ')
                self.current_dir[dirname] = {}
            else:
               size, filename = self.current_command.split(' ')
               self.current_dir[filename] = size

            self.cursor += 1


    def chdir(self):

        c = self.current_command

        c = c.replace('$ cd ', '')

        if c == '..':
            self.current_dir_list.pop()
        elif c == '/':
            self.current_dir_list = ['/']
        else:
            self.current_dir_list.append(c)
    
        self.chdir_dict()   
  

    def chdir_dict(self):

        self.current_dir = self.tree

        for item in self.current_dir_list:
            self.current_dir = self.current_dir[item]
        
```

```python
def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

```python
flat_tree = flatten(i.tree)
flat_tree
```

```python
dir_sizes = {key.rsplit('_', 1)[0]:0 for key in flat_tree.keys()}

def add_to_parents(dir_sizes:dict, path:str, size):

    parent = path.rsplit('_', 1)[0]
    if not parent == path:
        # print(f'Adding to {parent=} {path=}')
        _ = dir_sizes.get(parent, 0)
        dir_sizes[parent] = _ + int(size)
        add_to_parents(dir_sizes, parent, size)
    else:
        return


for key, val in flat_tree.items():
    add_to_parents(dir_sizes, key, val)

dir_sizes
    
```
