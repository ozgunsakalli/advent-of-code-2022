```python
with open('input-day1.txt', 'rt') as f:
    d = f.read()
```

```python
d_list = d.split('\n')
d_list
```

```python
d_cumsum = []
temp_sum = 0
for i, item in enumerate(d_list):

    if i == len(d_list) or item == '':
        d_cumsum.append(temp_sum)
        temp_sum = 0
    else:
        temp_sum += int(item)

d_cumsum
```

```python
# answer to part 1
max(d_cumsum)
```

Part 2

```python
top3 = sorted(d_cumsum)[-3:]
top3
```

```python
# answer to part 2
sum(top3)
```
